```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace('-', '_')
```

```python
print(df.columns)
```

```python
df['scheduledday'] = pd.to_datetime(df['scheduledday'])
df['appointmentday'] = pd.to_datetime(df['appointmentday'])
```

```python
print(df.columns)
```

```python
df = df[df['age'] >= 0]
```

```python
print(df.columns)
```

```python
df['waiting_days'] = (df['appointmentday'] - df['scheduledday']).dt.days
```

```python
df['appointment_dayofweek'] = df['appointmentday'].dt.dayofweek
df['scheduled_dayofweek'] = df['scheduledday'].dt.dayofweek
```

```python
df['no_show'] = df['no_show'].map({'No': 0, 'Yes': 1})
```

```python
df['no_show'].unique()
```

```python
df['no_show'].value_counts()
```

```python
df['no_show'].dtype
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
```

```python
df.drop(['patientid', 'appointmentid'], axis=1, inplace=True)
```

```python
print(df.columns)
```

```python
df.info()
```

```python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['age'] >= Q1 - 1.5*IQR) & 
        (df['age'] <= Q3 + 1.5*IQR)]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['age', 'waiting_days']] = scaler.fit_transform(df[['age', 'waiting_days']])
```

```python
print("Mean:\n", df[['age', 'waiting_days']].mean())
print("\nStandard Deviation:\n", df[['age', 'waiting_days']].std(ddof=0))
```

```python
print(df.head())
print(df.shape)

df.to_csv("cleaned_medical_dataset.csv", index=False)
```
